# Item Dictionary From Raw

Take raw list of mythological items from Wikipedia (https://en.wikipedia.org/wiki/List_of_mythological_objects) and convert to JSON object or clean csv

```python
import csv

data = []

with open('items_raw.csv', newline='') as csvfile:
    reader = csv.reader(csvfile, delimiter=',')
    for row in reader:
        data.append(row)
```

```python
# Create crude dictionary

item_dict = {}

def item_type(row_list):
    joined_row = ''.join(row_list)
    if len(joined_row.split(' ')) < 4:
        return True
    return False

def create_item(item_raw):
    item_name = ''.join(item_raw.split(',')[0])
    item_desc = clean_desc(''.join(item_raw.split(',')[1:]))
    return {item_name:item_desc}

def clean_desc(item_desc):
    item_desc = item_desc.strip("\\")
    return item_desc.strip()


for row in data:
    if item_type(row):
        current_header = ' '.join(row).lower()
    else:
        if current_header in item_dict.keys():
            item_dict[current_header].append(
                    create_item(''.join(row))
            )
        else:
            item_dict[current_header] = []
        
item_dict
```
